# k-means Clustering

```python
(m, c) = kmeans(img, 20, 30)
```

```python
plt.xlabel('Wavelenghth Band')
plt.ylabel('Reflectance')
for i in range(c.shape[0]):
    plt.plot(c[i])
plt.show()
```

```python
view = imshow(img, bands=TRUE_COLOR,stretch=0.01, classes=m)
view.set_display_mode('overlay')
view.class_alpha = 0.6
view.show_data
```

# Anlisis de Componentes Principales

```python
pc = principal_components(img)
```

```python
v = imshow(pc.cov)
```

```python
pc_0999 = pc.reduce(fraction=0.999)
```

```python
len(pc_0999.eigenvalues)
```

```python
img_pc = pc_0999.transform(img)
```

```python
v = imshow(img_pc[:,:,:3], stretch_all=True)
```

```python
classes = create_training_classes(img_pc, gt)
```

```python
gmlc = GaussianClassifier(classes)
```

```python
clmap = gmlc.classify_image(img_pc)
```
